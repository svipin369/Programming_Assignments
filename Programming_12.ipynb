```python
# Question 1- Program to extract unique values from a dictionary's values
def extract_unique_values(input_dict):
    unique_values = set()
    
    for values in input_dict.values():
        if isinstance(values, list):
            unique_values.update(values)
        else:
            unique_values.add(values)
    
    return list(unique_values)

# Input the number of key-value pairs from the user
try:
    num_pairs = int(input("Enter the number of key-value pairs: "))
except ValueError:
    print("Please enter a valid number.")
    exit(1)

# Input the dictionary from the user
input_dict = {}
for i in range(num_pairs):
    key = input(f"Enter key {i + 1}: ")
    value = input(f"Enter value for '{key}': ")
    try:
        value = int(value)  # Convert value to integer if possible
    except ValueError:
        pass  # Continue with the value as a string if it can't be converted
    input_dict[key] = value

# Call the function to extract unique values
unique_values = extract_unique_values(input_dict)

# Display the original dictionary
print("Original Dictionary:")
for key, value in input_dict.items():
    print(f"'{key}': {value}")

# Display the unique values
print("\nUnique values from the dictionary:", unique_values)
```

```python
# Question 2- Program to find the sum of all items in a dictionary
def sum_of_dictionary_items(input_dict):
    total_sum = sum(input_dict.values())
    return total_sum

# Input the number of key-value pairs from the user
try:
    num_pairs = int(input("Enter the number of key-value pairs: "))
except ValueError:
    print("Please enter a valid number.")
    exit(1)

# Input the dictionary from the user
input_dict = {}
for i in range(num_pairs):
    key = input(f"Enter key {i + 1}: ")
    value = input(f"Enter value for '{key}': ")
    try:
        value = int(value)  # Convert value to integer if possible
    except ValueError:
        print("Invalid value. Please enter a valid number.")
        continue
    input_dict[key] = value

# Call the function to find the sum of all items
result_sum = sum_of_dictionary_items(input_dict)

# Display the sum
print("Sum of all items in the dictionary:", result_sum)
```

```python
# Question 3- Program to merge two dictionaries
def merge_two_dicts(dict1, dict2):
    merged_dict = dict1.copy()
    merged_dict.update(dict2)
    return merged_dict

# Input the number of key-value pairs for the first dictionary
try:
    num_pairs_dict1 = int(input("Enter the number of key-value pairs for the first dictionary: "))
except ValueError:
    print("Please enter a valid number.")
    exit(1)

# Input the first dictionary from the user
dict1 = {}
for i in range(num_pairs_dict1):
    key = input(f"Enter key {i + 1} for the first dictionary: ")
    value = input(f"Enter value for '{key}' in the first dictionary: ")
    dict1[key] = value

# Input the number of key-value pairs for the second dictionary
try:
    num_pairs_dict2 = int(input("Enter the number of key-value pairs for the second dictionary: "))
except ValueError:
    print("Please enter a valid number.")
    exit(1)

# Input the second dictionary from the user
dict2 = {}
for i in range(num_pairs_dict2):
    key = input(f"Enter key {i + 1} for the second dictionary: ")
    value = input(f"Enter value for '{key}' in the second dictionary: ")
    dict2[key] = value

# Call the function to merge the two dictionaries
merged_dictionary = merge_two_dicts(dict1, dict2)

# Display the merged dictionary
print("Merged Dictionary:")
for key, value in merged_dictionary.items():
    print(f"'{key}': {value}")
```

```python
# Question 4- Program to convert a list of key-value pairs to a flat dictionary
def list_to_flat_dictionary(key_value_list):
    flat_dict = {}
    for i in range(0, len(key_value_list), 2):
        if i + 1 < len(key_value_list):
            key = key_value_list[i]
            value = key_value_list[i + 1]
            flat_dict[key] = value
    return flat_dict

# Input the number of key-value pairs from the user
try:
    num_pairs = int(input("Enter the number of key-value pairs: "))
except ValueError:
    print("Please enter a valid number.")
    exit(1)

# Input the list of key-value pairs one by one
key_value_list = []
for i in range(num_pairs):
    key = input(f"Enter key {i + 1}: ")
    value = input(f"Enter value for '{key}': ")
    key_value_list.extend([key, value])

# Call the function to convert the list to a flat dictionary
flat_dictionary = list_to_flat_dictionary(key_value_list)


print("Flat Dictionary:", flat_dictionary)
```

```python
#Question 5- Program to insert a key-value pair at the beginning of an OrderedDict

from collections import OrderedDict
def insert_at_beginning(original_dict, item):
    new_dict = OrderedDict()
    new_dict[item[0]] = item[1]
    new_dict.update(original_dict)
    return new_dict

# Input the number of key-value pairs from the user
try:
    num_pairs = int(input("Enter the number of key-value pairs in the original dictionary: "))
except ValueError:
    print("Please enter a valid number.")
    exit(1)

# Input the original dictionary from the user
original_dict = OrderedDict()
for i in range(num_pairs):
    key = input(f"Enter key {i + 1}: ")
    value = input(f"Enter value for '{key}': ")
    original_dict[key] = value

# Input the item to be inserted as a tuple
item_to_insert = input("Enter the item to insert as a tuple (e.g., 'key value'): ").split()
if len(item_to_insert) != 2:
    print("Invalid input for item. Please provide a key and a value.")
    exit(1)

# Call the function to insert the item at the beginning
result_dict = insert_at_beginning(original_dict, (item_to_insert[0], item_to_insert[1]))

# Display the result
print("Original Dictionary:", original_dict)
print("Modified Dictionary:", result_dict)

```

```python
#Question 6- Program to check the order of characters in a string
from collections import OrderedDict
def check_order(input_string, pattern):
    # Remove spaces from the string and convert it to lowercase
    clean_string = input_string.replace(" ", "").lower()

    # Initialize the position of each character in the pattern
    pattern_index = 0

    # Iterate through the characters in the cleaned input string
    for char in clean_string:
        if char == pattern[pattern_index]:
            pattern_index += 1
            if pattern_index == len(pattern):
                return True

    return False

# Input the string from the user
input_string = input("Enter a string: ")

# Input the pattern from the user
pattern = input("Enter a pattern to check: ")

# Call the function to check the order of characters
result = check_order(input_string, pattern)

# Display the result
if result:
    print(f"The pattern '{pattern}' appears in the string in order.")
else:
    print(f"The pattern '{pattern}' does not appear in the string in order.")
```

```python
#QUESTION 7- Program to sort Dictionary by key or value

# Function to sort a dictionary by its keys
def sort_dict_by_key(input_dict):
    return dict(sorted(input_dict.items()))

# Function to sort a dictionary by its values
def sort_dict_by_value(input_dict):
    return dict(sorted(input_dict.items(), key=lambda item: item[1]))

# Input a dictionary from the user
input_dict = {}
try:
    num_pairs = int(input("Enter the number of key-value pairs in the dictionary: "))
except ValueError:
    print("Please enter a valid number.")
    exit(1)

for i in range(num_pairs):
    key = input(f"Enter key {i + 1}: ")
    value = input(f"Enter value for '{key}': ")
    input_dict[key] = value

# Ask the user for sorting preference
sort_choice = input("Sort by key or value (enter '1' for key, '2' for value): ")

# Sort the dictionary based on the user's choice
if sort_choice == '1':
    sorted_dict = sort_dict_by_key(input_dict)
elif sort_choice == '2':
    sorted_dict = sort_dict_by_value(input_dict)
else:
    print("Invalid sorting choice. Please enter '1' for key or '2' for value.")
    exit(1)

# Display the sorted dictionary
print("Sorted Dictionary:")
for key, value in sorted_dict.items():
    print(key, value)
```
