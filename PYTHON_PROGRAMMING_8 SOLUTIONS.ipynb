```python
## Q1. Addition of Matrix ####

def add_matrices(matrix1, matrix2):
    rows = len(matrix1)
    columns = len(matrix1[0])
    result = [[0 for _ in range(columns)] for _ in range(rows)]

    for i in range(rows):
        for j in range(columns):
            result[i][j] = matrix1[i][j] + matrix2[i][j]

    return result

rows = int(input("Enter the number of rows for the first matrix: "))
columns = int(input("Enter the number of columns for the first matrix: "))

matrix1 = []
print("Enter the elements of the first matrix:")
for i in range(rows):
    row = []
    for j in range(columns):
        element = int(input("Enter element at position ({}, {}): ".format(i+1, j+1)))
        row.append(element)
    matrix1.append(row)

print("\nEnter the elements of the second matrix:")
matrix2 = []
for i in range(rows):
    row = []
    for j in range(columns):
        element = int(input("Enter element at position ({}, {}): ".format(i+1, j+1)))
        row.append(element)
    matrix2.append(row)

result_matrix = add_matrices(matrix1, matrix2)
print("\nResultant matrix after addition:")
for row in result_matrix:
    print(row)


```

```python
#### Q2. Multiplication of Matrix #####

def multiply_matrices(matrix1, matrix2):
    rows1 = len(matrix1)
    columns1 = len(matrix1[0])
    rows2 = len(matrix2)
    columns2 = len(matrix2[0])

    if columns1 != rows2:
        print("Invalid matrix dimensions for multiplication.")
        return None

    
    result_matrix = []
    for i in range(rows1):
        row = []
        for j in range(columns2):
            row.append(0)
        result_matrix.append(row)

    
    for i in range(rows1):
        for j in range(columns2):
            for k in range(columns1):
                result_matrix[i][j] += matrix1[i][k] * matrix2[k][j]

    return result_matrix
rows1 = int(input("Enter the number of rows for the first matrix: "))
columns1 = int(input("Enter the number of columns for the first matrix: "))

matrix1 = []
print("Enter the elements of the first matrix:")
for i in range(rows1):
    row = []
    for j in range(columns1):
        element = int(input("Enter element at position ({}, {}): ".format(i+1, j+1)))
        row.append(element)
    matrix1.append(row)

# Take input from the user for the second matrix
rows2 = int(input("Enter the number of rows for the second matrix: "))
columns2 = int(input("Enter the number of columns for the second matrix: "))

matrix2 = []
print("\nEnter the elements of the second matrix:")
for i in range(rows2):
    row = []
    for j in range(columns2):
        element = int(input("Enter element at position ({}, {}): ".format(i+1, j+1)))
        row.append(element)
    matrix2.append(row)

# Display the input matrices
print("\nFirst matrix:")
for row in matrix1:
    print(row)

print("\nSecond matrix:")
for row in matrix2:
    print(row)

# Multiply the matrices
result_matrix = multiply_matrices(matrix1, matrix2)

# Display the result
if result_matrix:
    print("\nResultant matrix after multiplication:")
    for row in result_matrix:
        print(row)
```

```python
#### Q3. Transpose of a Matrix ###

def transpose_matrix(matrix):
    rows = len(matrix)
    columns = len(matrix[0])
    transposed_matrix = []
    for j in range(columns):
        row = []
        for i in range(rows):
            row.append(matrix[i][j])
        transposed_matrix.append(row)

    return transposed_matrix
rows = int(input("Enter the number of rows for the matrix: "))
columns = int(input("Enter the number of columns for the matrix: "))

matrix = []
print("Enter the elements of the matrix:")
for i in range(rows):
    row = []
    for j in range(columns):
        element = int(input("Enter element at position ({}, {}): ".format(i+1, j+1)))
        row.append(element)
    matrix.append(row)
print("\nInput matrix:")
for row in matrix:
    print(row)
transposed_matrix = transpose_matrix(matrix)
print("\nTransposed matrix:")
for row in transposed_matrix:
    print(row)
```

```python
#### Q4. Sort Words in Alphabetic Order ####

def sort_words(words):
    sorted_words = sorted(words)
    return sorted_words

total_words = int(input("Enter the number of words: "))
words_list = []
for i in range(total_words):
    word = input("Enter word {}: ".format(i+1))
    words_list.append(word)
sorted_words = sort_words(words_list)
print("Words in alphabetical order:")
for word in sorted_words:
    print(word)
```

```python
##### Q5. Remove Punctuation From a String #####

import string
def remove_punctuation(input_string):
    punctuation = string.punctuation
    no_punct = ""
    for char in input_string:
        if char not in punctuation:
            no_punct += char
    return no_punct

input_string = input("Enter a string: ")
processed_string = remove_punctuation(input_string)
print("String without punctuation:", processed_string)
```
